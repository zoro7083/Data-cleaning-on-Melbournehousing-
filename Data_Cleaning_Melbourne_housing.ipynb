### Summing up all null values column-wise

```python
data.isnull()  # It will return True where value is Null
data.isnull().sum()   # It will return the sum of all Null values column-wise
data.isnull().sum().sort_values(ascending = False)   # Sorting Values in Descending Order
```

```python
data.isnull()  # It will return True where value is Null
data.isnull().sum(axis=1)   # It will return the sum of all Null values column-wise
data.isnull().sum(axis=1).sort_values(ascending = False)   # Sorting Values in Descending Order
```

### Showing Columns & Rows having all null values

### Showing Columns and Rows having atleast one null value

### Summing up the missing values (column-wise) - in %

```python
data.isnull().sum().sort_values(ascending = False)/len(data) * 100
round(data.isnull().sum().sort_values(ascending = False)/len(data) * 100, 2)
```

### Describe Longtitude and Lattitude and later imputing the values by mean

```python
data['Longtitude'].describe()
```

```python
data['Lattitude'].describe()
```

```python
data.loc[:,['Lattitude','Longtitude']].describe()
```

```python
data['Longtitude'].mean()
```

```python
data['Lattitude'].mean()
```

```python
data.loc[:,'Lattitude'].fillna(data['Lattitude'].mean(),inplace=True)
```

```python
data.loc[:,'Longtitude'].fillna(data['Longtitude'].mean(),inplace=True)
```

```python
round(data.isnull().sum().sort_values(ascending=False)/len(data) * 100, 2)
```

### Replacing Null Values from Bathroom & Bedroom2

```python
data.loc[:,['Bathroom', 'Bedroom2']].describe()
```

```python
data.groupby('Bathroom').Bathroom.count().sort_values(ascending=False)
# data['Bathroom'].astype('category').value_counts()
```

```python
data.groupby('Bedroom2').Bedroom2.count().sort_values(ascending=False)
# data['Bedroom2'].astype('category').value_counts()
```

```python
data.loc[:,'Bathroom'].fillna(1,inplace=True)
```

```python
data.loc[:,'Bedroom2'].fillna(3,inplace=True)
```

```python
round(data.isnull().sum().sort_values(ascending=False)/len(data) * 100, 2)
```

### Replacing Null Values from Car

```python
data['Car'].describe()
```

```python
data.groupby('Car').Car.count().sort_values(ascending=False)
```

```python
data.loc[:,'Car'].fillna(2,inplace=True)
```
